# 3.리눅스 용어

- pwd : 현재 위치
- ls : 현재 위치에 있는 파일이나 디렉터리
    - ls -l : 상세 표시까지 함
    - ls -a : 숨긴 파일도 함께
    - ls -r : 파일 순서 거꾸로
    - ls -t : 시간순 정렬
- clear : 화면 정리
- cd .. : 뒤로
- cd 이름 : 이름으로 이동
- cd ~ : 홈으로
- mkdir 폴더이름 : 폴더(디렉터리) 생성
- rm -r 폴더이름 : 폴더(디렉터리) 삭제
- vim 이름.txt : 메모장 생성
    - 메모장 생성 후 : 'a' 누르면 입력 모드/ 'Esc' 누르면 저장 종료 모드
    - 'a' 모드 할 때 반드시 영어키로
    - ':wq' : 저장 종료 
- cat 파일 : 파일 내용 알려줌

# 4.버전 관리

## 1) 깃 저장소 만들기

![png](images/1.PNG)

- 작업 트리 : 눈에 보이는 디렉터리
- .git 디렉터리 : 해당 디렉터리에 들어가서 git init으로 생성해준다. 눈에 안보임
- 스테이지 : 버전을 만들기 위해 대기하는 곳
- 저장소 : 스테이지에 있던 파일을 버전으로 만들어 저장하는 곳(커밋 명령으로)

- git status

![image.png](images/2.PNG)

스테이지에 올리기

- git add 파일이름
- 그리고 git status 하면 new file에 보임

커밋하기

- git commit -m "message1"
- -m 은 메시지 추가하는 옵션
- git status 로 clean 확인
- git log 로 기록 확인
    - git log --oneline : 한 줄에 하나씩

스테이징 커밋 한 번에

- git commit -am "message2"

커밋 전에 수정하기

- git checkout -- 파일이름

스테이징 되돌리기

- git reset HEAD 파일이름
- git reset HEAD^ : 최신 커밋 되돌리기
- git reset --hard 보사한 커밋 해시 : 특정 커밋으로 되돌리기

깃에서 로컬로 파일 복제

- 깃허브에서 복제할 파일 들어가 clone or download 에서 주소 복사
- git clone 복사한 주소 doit

## 2)브랜치

기본

- git branch : 깃에서 브랜치를 만들거나 확인
    - master는 기본
    - git branch 이름 : 새로운 브랜치 생성
- git checkout 브랜치이름 : 브랜치 이동
- git log --oneline --branches : 한 줄씩 각 브랜치마다
- git log --oneline --branches --graph : 그래프도 보여줌/ 부모에서 파생되는 걸 점선으로 보여줌

브랜치끼리 비교

- git log 브랜치1..브랜치2 : 1에는 없고 2에 있는거 보여줌

병합하기

- master로 체크아웃 한 상태에서
- git merge 병합할 브랜치 이름
- 병합하면 수정 내용의 위치가 다를 경우 전부 합쳐줌
- 수정한 부분이 같은 상태에서 병합할 경우 ==== 표시로 모두 다 나옴(수정내용을 보고 직접 수정해줘야 함)

브랜치 삭제

- git branch -d 브랜치이름

브랜치 이동

- 해당 브랜치로 체크아웃 한 후
- git reset 커밋 해시 : 해시로 브랜치 이동, 이전에 있던 커밋 삭제됨

# 5.백업하기

기능

- 지금까지 한 작업은 local 저장소에 저장한 것임
- 깃허브: 백업을 위한 온라인 저장소
- 깃허브는 무료 유료 다 있음/ 깃 설치 안해도 온라인상 사용가능
- 다른 사람의 소스를 가져와 나름대로 수정해 볼 수 있음

1. [가입 먼저 하기](https://github.com/)
- 오른쪽 상단 + 누르고 저장소 만들기
    - 저장소 이름은 : https://github.com/유저이름/저장소이름.git

로컬 저장소 -> 깃허브

1. 접속
    - git remote add origin 저장소주소
    - 오류가 없다면 잘 실행된 것
    - 확인 방법: git remote -v
- 옮기기
    - git push -u origin master
    - 로그인하기
    - 한 번 푸시했던 저장소면(연결된 저장소) git push만 해도 됨
- 깃허브에서 파일 내려받기
    - 저장소 접속 후
    - git pull origin master // 깃허브(origin)에서 브랜치(master)로 가져오겠다는 뜻

SSH로 접근하기

- 그냥 로그인하는 방식도 있는데
- 보안이 더 좋음
- 로그인 계속하는 번거로움이 없음

1. ssh-keygen하고 엔터 세 번해서 확인
- cd ~/.ssh 로 퍼블릭이랑 프라이빗 키 있는 폴더로 이동
- ls -la 로 확인
- cat id_rsa.pub 으로 키 불러오기
- 키 복사 해서 깃에 등록
    - setting에 SSH 머시기 들어가서 등록

접속 방법

- cd ~
- git init connect-ssh
- cd connect-ssh
- git remote add origin 주소넣기(ssh)
