```python
import plotly.offline as offline
import plotly.graph_objs as go
```

```python
offline.init_notebook_mode(connected=True)

trace = go.Candlestick(x=data['id'],
                      open = data['open'],
                      high = data['high'],
                      low = data['low'],
                      close = data['close'])

d = [trace]

layout = go.Layout(title = "Paycoin Candle Chart")
fig = go.Figure(data = d, layout = layout)
offline.iplot(fig,filename = "candlestick")
```

```python
dd = [
    go.Bar(
        x=data['id'],
        y=data['amount']
    )
]
layout = go.Layout(title = "Paycoin Amount Bar Chart")
fig = go.Figure(data = dd, layout = layout)
offline.iplot(fig,filename = "bar chart")
```
